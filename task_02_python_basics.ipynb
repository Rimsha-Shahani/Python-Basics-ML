```python
# Variables and Data Types

name = "Rimsha"
print(name)

age = 20
print(age)

cgpa = 3.7
print(cgpa)

is_student = True
print(is_student)

subjects = ["AI", "ML", "Python"]
print(subjects)

marks = (80, 85, 90)
print(marks)

student = {
    "Name": "Rimsha",
    "Age": 20
}
print(student)

# For Loop

for i in range(1, 6):
    print(i)

# While Loop

count = 1

while count <= 3:
    print(count)
    count += 1

# If Else

number = 10

if number > 5:
    print("Greater than 5")
else:
    print("Less than 5")

# Functions

def square(x):
    return x * x

print(square(5))

def add(a, b):
    return a + b

print(add(10, 20))

# List Comprehension

squares = [x**2 for x in range(1, 6)]
print(squares)

# NumPy

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(arr)
print(arr.mean())

# Pandas

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha"],
    "Age": [20, 21, 22, 23],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print(df)

df.info()

print(df.head())

print(df.describe())
```
